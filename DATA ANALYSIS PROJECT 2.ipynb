```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Dataset  (1).csv')  


rating_column = 'Aggregate rating'
votes_column = 'Votes'

plt.figure(figsize=(10, 6))
df[rating_column].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()

most_common_rating_range = pd.cut(df[rating_column], bins=[0, 2, 4, 6, 8, 10], include_lowest=True).value_counts().idxmax()
print("\nMost Common Rating Range:")
print(most_common_rating_range)

average_votes = df[votes_column].mean()
print("\nAverage Number of Votes Received by Restaurants:", average_votes)
```

```python
import pandas as pd
from collections import Counter


df = pd.read_csv('Dataset  (1).csv')  


cuisines_column = 'Cuisines'
rating_column = 'Aggregate rating'

cuisine_combinations = df[cuisines_column].str.split(', ')
flattened_cuisines = [cuisine for sublist in cuisine_combinations.dropna() for cuisine in sublist]
common_cuisine_combinations = Counter(flattened_cuisines).most_common(10)

print("\nMost Common Combinations of Cuisines:")
for combination, count in common_cuisine_combinations:
    print(f"{combination}: {count} times")


average_ratings_by_cuisine_combination = df.groupby(cuisines_column)[rating_column].mean()


top_rated_cuisine_combinations = average_ratings_by_cuisine_combination.sort_values(ascending=False).head(10)
print("\nTop 10 Cuisine Combinations with Highest Average Ratings:")
print(top_rated_cuisine_combinations)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Dataset  (1).csv')  


restaurant_name_column = 'Restaurant Name'
rating_column = 'Aggregate rating'
votes_column = 'Votes'

restaurant_chains = df[restaurant_name_column].value_counts()
chains_present = restaurant_chains[restaurant_chains > 1]

print("\nRestaurant Chains Present in the Dataset:")
print(chains_present)


average_rating_by_chain = df.groupby(restaurant_name_column)[rating_column].mean()
popularity_by_chain = df.groupby(restaurant_name_column)[votes_column].sum()


top_rated_chains = average_rating_by_chain.sort_values(ascending=False).head(10)
print("\nTop 10 Restaurant Chains by Average Rating:")
print(top_rated_chains)


most_popular_chains = popularity_by_chain.sort_values(ascending=False).head(10)
print("\nTop 10 Restaurant Chains by Popularity:")
print(most_popular_chains)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
top_rated_chains.plot(kind='bar', color='skyblue')
plt.title('Top 10 Chains by Average Rating')
plt.xlabel('Restaurant Chain')
plt.ylabel('Average Rating')

plt.subplot(1, 2, 2)
most_popular_chains.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Chains by Popularity')
plt.xlabel('Restaurant Chain')
plt.ylabel('Total Votes')

plt.tight_layout()
plt.show()

```
